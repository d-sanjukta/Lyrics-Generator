### HUMAN EVALUATION:
Here, we will perform Exploratory Data Analysis (EDA) on the data collected using a Google Form. The form was designed to gather responses for two separate lyrics generated by the GPT-2 and DistilGPT-2 models based on a common prompt.

Our objective is to analyze the collected data in a more organized and meaningful manner.

> Importing the image

RESPONSE SHEET

![Human Evaluation](https://drive.google.com/uc?id=1Eyk5AKUd7N_sL7E-xrypnXm3X_J_HBQc)

* We have a total of 24 responces out of which (6 examples in each response):
  * 76 is for Prompt 1 (GPT2 model)
  * 68 is for Prompt 2 (distill GPT2 model)

```python
import matplotlib.pyplot as plt

# Lets make a simple bar plot for the above results :

# Vote counts for Prompt 1 and Prompt 2
vote_counts = {'Prompt 1\nGPT2': 76, 'Prompt 2\ndistil_GPT2': 68}

# Extract the prompts and vote counts as separate lists
prompts = list(vote_counts.keys())
votes = list(vote_counts.values())

# Create a bar plot
plt.bar(prompts, votes, width = 0.2, color = 'purple')

# Add labels and title
plt.xlabel('Prompts', size= 15)
plt.ylabel('Vote Counts', size = 15)
plt.title('Vote Counts for Prompts', size = 15, fontweight  = 'bold')

# Display the plot
plt.show()
```

> INFERENCE:

* Upon thorough analysis, it can be concluded that while GPT-2 received a higher number of votes, there is no significant disparity in the distribution of votes between Prompt 1 and Prompt 2. However, as previously inferred, the lyrics generated by GPT-2 exhibit better coherence and rhythm compared to DistilGPT-2.

* Based on this observation, the decision has been made to proceed with the GPT-2 model for training the remaining datasets and fine-tuning the artist-specific models. This choice is driven by the superior performance and lyrical quality demonstrated by GPT-2 in the initial evaluation.
